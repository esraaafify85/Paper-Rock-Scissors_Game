```python
# "Paper - Rock - Scissors Game" using function, if, else and elif.
###################################################################

'''

** Game Description **
-------------------------------------------------------------------------------------------------------
- Ask for player plays (using input), 
- Compare them with the random computer input, 
- Print out message of congratulations to the winner, and
- Ask if the players want to start a new game.
-------------------------------------------------------------------------------------------------------

** Game rules**
-------------------------------------------------------------------
Paper    beats  Rock     : Paper wins because paper covers rock.
Paper    beats  Scissors : Scissors win because scissors cut paper.
Scissors beats  Rock     : Rock wins because rock smashes scissors.
Everything else is a tie.
-------------------------------------------------------------------

** Game probabilities **
------------------------------------
p | p = p --> Tie, please try again.
r | r = r --> Tie, please try again.
s | s = s --> Tie, please try again.
----------------------------------------------------------------
p | r = p --> You win, paper covers rock, Good job!
p | S = s --> Computer win, scissors cut paper!
----------------------------------------------------------------
r | p = p --> Computer win, paper covers rock!
r | s = r --> You win, rock smashes scissors, Good job!
----------------------------------------------------------------
s | p = s --> You win, scissors cut paper, Good job!
s | r = r --> Computer win, rock smashes scissors!
----------------------------------------------------------------

-- >   if the player enter wrong choice or any number else like "3, 5 or 100 ..etc."  
       this msg'll be printed --> "Wrong Choice, please try again."
       
'''

#------------------------------------------------------------------------------
# Starting the game code:
#------------------------------------------------------------------------------
# Welcome message for player.
print("Hello! Welcome to the Paper-Rock-Scissors game!")
print("Let's start the game!")
print('\n')


# define p_r_c_game() function.
#----------------------------------------------------------------------------------------------------
def p_r_s_game():   

    # use import random library for computer choices.
    from random import randint   

    # ask player and computer for choices.
    player = int(input("For Paper enter 0 , Rock enter 1, Scissors enter 2: "))
    computer = randint(0,2)

#-----------------------------------------------------------------
# tie choices:
#-----------------------------------------------------------------
# player choose "paper" vs computer choose "paper".
    if player == 0 and computer == 0:
        print('You choose "paper" and computer choose "paper".')
        print('Tie, please try again.')  

# player choose "rock" vs computer choose "rock".
    elif player == 1 and computer == 1:
        print('You choose "rock" and computer choose "rock".')
        print('Tie, please try again.')

# player choose "scissors" vs computer choose "scissors".  
    elif player == 2 and computer == 2:
        print('You choose "scissors" and computer choose "scissors".')
        print('Tie, please try again.')

#----------------------------------------------------------------- 
# paper vs rock & scissors choices:
#-----------------------------------------------------------------
# player chooce "paper" vs computer choose "rock".
    elif player == 0 and computer == 1:
        print('You choose "paper" and computer choose "rock".')
        print('You win, paper covers rock, Good job!')

# player chooce "paper" vs computer choose "scissors".
    elif player == 0 and computer == 2:
        print('You choose "paper" and computer choose "scissors".')
        print('Computer win, scissors cut paper!')

#-----------------------------------------------------------------
# rock vs paper & scissors choices:
#----------------------------------------------------------------- 
# player chooce "rock" vs computer choose "paper".
    elif player == 1 and computer == 0:
        print('You choose "rock" and computer choose "paper".')
        print('Computer win, paper covers rock!')

# player chooce "rock" vs computer choose "scissors".
    elif player == 1 and computer == 2:
        print('You choose "rock" and computer choose "scissors".')
        print('You win, rock smashes scissors, Good job!')

#-----------------------------------------------------------------
# scissors vs paper & rock choices:
#----------------------------------------------------------------- 
# player chooce "scissors" vs computer choose "paper".   
    elif player == 2 and computer == 0:
        print('You choose "scissors" and computer choose "paper".')
        print('You win, scissors cut paper, Good job!') 

# player chooce "scissors" vs computer choose "rock".
    elif player == 2 and computer == 1:
        print('You choose "scissors" and computer choose "rock".')
        print('Computer win!, rock smashes scissor!')

#-----------------------------------------------------------------
#Wrong Choice.
#-----------------------------------------------------------------
    else:
        print('Wrong Choice, please try again!') 
        print('Please enter for Paper 0 , Rock 1, Scissors 2.')

    # add playAgain() function to p_r_s_game() function.
    playAgain()

        
# define playAgain() function.
#------------------------------------------------------------------------------
def playAgain():
    
    # ask if the player wants to start a new game.
    print('\n')
    player_answer = input("Do you want to play again? Please type (YES or NO).")
    print('\n')

    # using the str.upper() to accept 'yes' or 'YES'.
    if player_answer.upper() == 'YES':
        print("OK, let's start a new game!")
        
        # call the p_r_s_game() function to start a new game.
        p_r_s_game() 
    
    # using the str.upper() to accept 'no' or 'NO'.
    elif player_answer.upper() == 'NO':
        print('Bye, see you in next game!')
        
    else:
        print("Wrong answer! Please type (YES or NO).")
        playAgain()


# call p_r_s_game() outside of the function.       
p_r_s_game() 
```
